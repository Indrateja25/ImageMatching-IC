```python
from PIL import Image
import pytesseract as tess
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('./../img1.jpeg') 
img.shape
```

```python
height = img.shape[0]
width = img.shape[1]

source_points = np.float32([[209, 191], [67, 1060],[751, 1155], [869, 286]])
destination_points = np.float32([[0, 0], [0, height],[width, height], [width, 0]])
homography_matrix = cv2.getPerspectiveTransform(source_points, destination_points)
source_points, destination_points, homography_matrix
```

```python
rectified_image1 = cv2.warpPerspective(img, homography_matrix, (width, height))
plt.imshow(rectified_image1)
```
