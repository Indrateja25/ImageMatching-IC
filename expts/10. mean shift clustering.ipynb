```python
def random_clustering(n, max_samples=10000):
    a = np.random.rand(n,5,4)
    d,l = {}, len(a)
    for i in range(l):
        key = list(a[i].reshape(-1))
        d[tuple(key)] = i
    idx = list(d.keys())
    ms = MeanShift(n_jobs=-1)
    for i in range(0, n, max_samples):
        minibatch = idx[i:i+max_samples]
        ms.fit(minibatch)
        print(i)
    clusters = []
    for p in idx:
        p1 = [np.array(p)]
        cluster = ms.predict(p1)
        clusters.append(cluster)
    return clusters
```
