```python
path = './'
vc = 1
for file in os.listdir("."):
    if(".MP4" in filename):
        continue
    video = cv2.VideoCapture(path+file)
    fc = 1
    try :
        while video.isOpened():
            frame_name = './frames/Video_'+str(vc)+'_Frame_'+str(fc)+'.png'
            _,img = video.read()
            cv2.imwrite(frame_name,img)
            print(frame_name)
            fc += 1
        video.release()
    except Exception as e :
        print(e)
    vc += 1
```

## Camera Calibration

```python
num_frames = 100
rows = 10
cols = 10
obj_points = []
img_points = []
```

```python
for i in range(500,500+num_frames):
    if(i%10 == 0):
        print(i)
    img = cv2.imread(f"./frames/frame{i}.png")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    corners = cv2.findChessboardCorners(gray, (rows, cols), None)
    print(corners)
    if corners is not None:
        obj_points.append(np.zeros((rows * cols, 3), np.float32))
        for j in range(rows):
            for k in range(cols):
                obj_points[-1][j * cols + k] = [k * 0.025, j * 0.025, 0]
        img_points.append(corners)
```
